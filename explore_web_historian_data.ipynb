# Researching your own autocomplete behavior
Web historian data can show you how you use autocomplete

```python
import pandas as pd
#df = pd.read_json("/content/web_historian_data_demo.json")
#df = pd.read_json("/content/web_historian_data.json")
df = pd.read_json("web_historian_data_demo.json")
df.head()
```

```python
#extract the oq parameter q from the url in df
df['oq'] = df['url'].str.extract(r'[&\?]oq=([^&\?]*)')
df['oq'] = df['oq'].str.replace(r'%20', ' ')
df['oq'] = df['oq'].str.replace(r'\+', ' ')
df.head()
```

```python
# find where searchTerms not equal oq and domain is google.com - show in a nice table
df_autocompletes = df[(df['searchTerms'] != df['oq']) & (df['oq'].notnull() == True) & (df['domain'] == 'google.com')]
df_autocompletes = df_autocompletes[['date','oq','searchTerms']]
df_autocompletes
```

```python
#export this file 
df_autocompletes.to_csv('autocomplete_data.csv', index=False)
```
